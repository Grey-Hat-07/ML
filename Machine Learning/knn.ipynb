```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
from sklearn.datasets import make_classification
import matplotlib.pyplot as plt
from sklearn.metrics import accuracy_score
from sklearn.neighbors import KNeighborsClassifier
```

```python
X, y = make_classification(
    n_samples=1000,
    n_features=2,
    n_informative=2,
    n_redundant=0,
    n_classes=2,
    random_state=42,
    n_clusters_per_class=1
)

X = pd.DataFrame(X)
X
```

```python
X = MinMaxScaler().fit_transform(X)
X = pd.DataFrame(X)
```

```python
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y)
```

```python
x_train,x_test,y_train,y_test = train_test_split(X, y, random_state=42, test_size=0.3)
x_test
```

```python
def euclidianDistance(x1, x2, y1, y2):
    dist = np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
    return dist
```

```python
# x_test = pd.DataFrame(x_test)
# x_train = pd.DataFrame(x_train)
```

```python
x_test = np.array(x_test)
x_train = np.array(x_train)
```

```python
distanceList = []
def calcPointDistace():
    for i in range(len(x_test)):
        individualDistList = []
        for j in range(len(x_train)):
            individualDistList.append(euclidianDistance(x_test[i][0], x_train[j][0], x_test[i][1], x_train[j][1]))
        distanceList.append(individualDistList)
calcPointDistace()
```

```python
distanceList = pd.DataFrame(distanceList)
distanceList
```

```python
combinedList = []
def combineClassWithDistance():
    for i in range(distanceList.shape[0]):
        combine = []
        for j in range(distanceList.shape[1]):
            combine.append((distanceList.iloc[i,j], y_train[j]))
        combine.sort(key=lambda x: x[0])
        combinedList.append(combine)
combineClassWithDistance()

combinedList = pd.DataFrame(combinedList)
combinedList
```

```python
k = int(input('Enter value of k: '))

sortedKList = []

for i in range(combinedList.shape[0]):
    kList = []
    for j in range(k):
        kList.append(combinedList.iloc[i,j])
    sortedKList.append(kList)

sortedKList = pd.DataFrame(sortedKList)
sortedKList
```

```python

categoryList = []
for i in range(sortedKList.shape[0]):
    ones = 0
    zeroes = 0
    for j in range(sortedKList.shape[1]):
        if sortedKList.iloc[i,j][1] == 1:
            ones += 1
        else: zeroes += 1
    if ones > zeroes:
        categoryList.append(1)
    else: categoryList.append(0)
categoryList
```

```python
np.sum(np.equal(y_test, categoryList))/len(y_test) * 100
```

```python
accuracy_score(y_test, categoryList)
```

```python
model =  KNeighborsClassifier().fit(x_train, y_train)
y_pred = model.predict(x_test)
accuracy_score(y_test, y_pred)
```
