```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
X, y = make_classification(
    n_samples=1200,
    n_features=2,
    n_classes=2,
    n_informative=2,
    n_redundant=0,
    n_clusters_per_class=1,
    n_repeated=False,
    random_state=42,
)

X = pd.DataFrame(X)
X.head()
```

```python
x_scalar = MinMaxScaler().fit_transform(X)
```

```python
plt.scatter(pd.DataFrame(x_scalar).iloc[:, 0], pd.DataFrame(x_scalar).iloc[:, 1], c=y)
```

```python
pd.DataFrame(X).plot(kind='hist')
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scalar, y, random_state=42, test_size=0.3)
x_test
```

```python
def euclidianDistance(x1, x2, y1, y2):
    return np.sqrt((x1 - x2) ** 2 + (y1 - y2) ** 2)
```

```python
len(x_test)
```

```python
individualDistance = []
def calcIndividualDistance():

    for i in range(len(x_test)):
        eachPointDistance = []
        for j in range(len(x_train)):
            eachPointDistance.append(euclidianDistance(x_test[i][0], x_train[j][0], x_test[i][1], x_train[j][1]))
        individualDistance.append(eachPointDistance)
calcIndividualDistance()

individualDistance = pd.DataFrame(individualDistance)
individualDistance
```

```python
len(y_train)
```

```python
LabelWithDistances = []
def mergeLabelsWithDistances():

    for i in range(individualDistance.shape[0]):
        addLabel = []
        for j in range(individualDistance.shape[1]):
            addLabel.append((individualDistance.iloc[i,j], y_train[j]))
        LabelWithDistances.append(addLabel)
mergeLabelsWithDistances()

LabelWithDistances = pd.DataFrame(LabelWithDistances)
LabelWithDistances
```

```python
sortData = []
def sortCombinedList():
    for i in range(LabelWithDistances.shape[0]):
        sortList = []
        for j in range(LabelWithDistances.shape[1]):
            sortList.append(LabelWithDistances.iloc[i,j])
        sortList.sort(key=lambda x: x[0])
        sortData.append(sortList)
sortCombinedList()

sortData = pd.DataFrame(sortData)
sortData
```

```python
k = int(input("Enter the value of k: "))

kSortedList = []
for i in range(sortData.shape[0]):
    kSortedList.append(sortData.iloc[i, :k])

kSortedList = pd.DataFrame(kSortedList)
kSortedList
```

```python

categoryList = []
for i in range(kSortedList.shape[0]):
    ones = 0
    zeroes = 0
    for j in range(kSortedList.shape[1]):
        if kSortedList.iloc[i,j][1] == 1:
            ones += 1
        else: zeroes += 1
    if ones > zeroes:
        categoryList.append(1)
    else: categoryList.append(0)
categoryList
```
